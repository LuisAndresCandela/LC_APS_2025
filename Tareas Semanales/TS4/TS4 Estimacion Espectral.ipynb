# TS4 : Primeras nociones de estimacion espectral

## Introcuccion


En esta TS vamos a adentrarnos primeramente en la representacion espectral de una señal.

Para ello, vamos a querer generar una señal de la cual no tenemos la certeza absoluta de en que frecuencia va a estar, es decir, que tenemos una incertidumbre de sus valores, lo cual vamos a representar con un variacion uniforme de dicha frecuencia. Esto nos ayuda a tener una situacion mas realista en la cual no vamos a saber a ciencia cierta el valor de la frecuencia de la señal con la que estamos trabajando y si un intervalo en el cual deberiamos esperar a la misma. 
Al igual que anteriormente, vamos a agregarle ruido a la señal de forma que estamos teniendo en cuenta los fenomenos que ocurren al muestrear una señal y analizarla posterior a esto. Como antes se menciona, esto es un paso mas que nos acerca a situaciones mas realistas en el mundo de la simulacion de señales.

La señal a representar va a estar dada por: 

$$ x(n) = a_1 sen(\Omega_1 * n ) + n_a(n) $$

Siendo:

$$ \Omega_1 = \Omega_0 + f_r \frac{2*\pi}{N} $$

$$ \Omega_0 = \frac{\pi}{2} $$

A su vez las variables aleatorias que tenemos estan dadas por:

$$ f_r \approx \mathcal{U}(-\frac{1}{2};\frac{1}{2}) $$

$$ n_a \approx \mathcal{N}(0;\sigma^2) $$

Con todo esto, como antes mencionamos queremos representar a la señal espectralmente pero a su vez no nos queremos quedar solo con esto sino evaluar que tan buena es nuestra representacion mediante el uso de estimadores. Estos son totalmente arbitrarios y los definimos en base a lo que estamos representando para asi poder nosotros mismos analizar la calidad de nuestro trabajo. Estos estimadores van a ser sobre $a_1$ y $\Omega_1$ y estan dados por:

$$
a^{i1} = \left| X_{iw}(\Omega_0) \right| = \left| \mathcal{F}\{x(n) \cdot w_i(n)\} \right|
$$

$$
\Omega^{i1} = \arg\max_{\Omega} \left\{ \left| X_{iw}(\Omega) \right| \right\}
$$







### Caracteristicas de las estimaciones

Se analizan distintas ventanas espectrales aplicadas a una señal sinusoidal a representar espectralmente:

- **Rectangular** (sin ventana)
- **Flattop**
- **Blackman-Harris**
- **Otras** ventanas que hayan sido previamente analizadas

### Condiciones del experimento:

```python
#%% módulos y funciones a importar

import numpy as np
from scipy import signal
from scipy.signal.windows import hamming, hann, blackman, kaiser, flattop, blackmanharris
from scipy.fft import fft, fftshift
import matplotlib as mpl
import matplotlib.pyplot as plt


#%% Genero mi Omega 1 que va a ser una frecuencia original mas una inceretidumbre en la frecuencia
# Esto genera que no sepa la frecuencia que estamos, sino que solo sepamos entre que valores va a estar

#%% Datos de la simulacion

fs = 1000           # frecuencia de muestreo (Hz)
N = 1000            # cantidad de muestras
ts = 1/fs           # tiempo de muestreo
df = fs/N           # resolución espectral

N_Test = 200        # Numero de pruebas

SNR = 10            # Signal to Noise Ratio 

Sigma2 = 10**-1     # Despejando el PN de la ecuacion de SNR llego a este valor

Omega_0 = fs/4      # Nos ponemos a mitad de banda digital

#%% Genero mi vector de 1000x200 de la senoidal

A1 = np.sqrt(2)                                             # Genero la amplitud de manera que el seno quede normalizado

fr = np.random.uniform(-1/2,1/2,N_Test).reshape(1,N_Test)   # Declaro mi vector 1x200 al pasar 200 valores a la uniforme     
                                                            # Fuerzo las dimensiones
Omega_1 = Omega_0 + fr*df                                   # Genero mi Omega 1

# Genero vector de tiempo para poder meterlo como mi vector de 1000 x 200 en el seno 
tt = np.linspace(0, (N-1)*ts, N).reshape(N,1)

tt = np.tile(tt, (1, N_Test))                               # Genero la matriz de 1000 x 200

# Al mutiplicar con * hacemos el producto matricial para que quede de 1000x200
S = A1 * np.sin(2 * np.pi * Omega_1 * tt)

varianzas = np.var(S, axis=0)                               # Varianza por cada una de las 200 señales
varianza_promedio = np.mean(varianzas)                      # Compruebo varianza 1 


#%% Genereo el ruido para la señal

# Para poder general la señal de ruido, tenemos que tener una distribucion normal con un N(o,sigma)

Media = 0                   # Media
SD_Sigma = np.sqrt(Sigma2)  # Desvio standar 

nn = np.random.normal(Media, SD_Sigma, N).reshape(N,1)              # Genero señal de ruido

nn = np.tile(nn, (1,N_Test))                                        # Ahora tengo que generaer mi matriz de ruido de 200x1000


#%% Ahora genero mi señal final sumando las matrices

# Esto seria mi x(k) = a1 * sen(W1 * k ) + n(k), siendo N(k) el ruido
Signal = S + nn

#%% Calculo las FFT de cada versión ( ventana, Hann, FlatTop, Blackman-Harris   )

#Declaro las ventanas que voy a usar y le doy N valores. Fuerzo dimensiones

ventanas = {
    'Rectangular': np.ones((N, 1)),
    'Hann': hann(N).reshape(N, 1),
    'FlatTop': flattop(N).reshape(N, 1),
    'Blackman-Harris': blackmanharris(N).reshape(N, 1)
}

num_senales = 200                                               # Cantidad de realizaciones
indices = np.linspace(0, N_Test - 1, num_senales, dtype=int)

# Eje de frecuencia solo parte positiva (0 a fs/2)
frec_pos = np.linspace(0, fs/2, N//2)

plt.figure(figsize=(12, 8))

# Ahora con mi vector ventana puedo ir generando cada una de las señales ventaneadas

for nombre, ventana in ventanas.items():
    
    S_windowed = Signal * ventana                   # Aplicamos la ventana

    X_f = fft(S_windowed, axis=0)                   # FFT y nos quedamos con la mitad positiva
    X_f = X_f[:N//2, :]  

    # Normalización por cada señal para llevar el pico a 0 dB
    X_f_norm = X_f / np.max(np.abs(X_f), axis=0)

    # Graficamos promedio de espectros en dB 
    espectro_promedio_db = 20 * np.log10(np.mean(np.abs(X_f_norm), axis=1) + 1e-12)
    
    plt.plot(frec_pos, espectro_promedio_db, label=nombre)

plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud |X(f)| [dB]")
plt.title("Comparación de ventanas - Parte positiva del espectro (Normalizado)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

#%% Estimador a1

# Acá estamos calculando cada uno de los estimadores para el promedio de todas las señales que tenemos en N/4
# En esto queremos tener la menor dispersion posible, por lo que, queremos que nuestro histograma este lo mas comprimido posible
# Y si vemos el grafico, como todas estan masomenos centradas en le mismo lugar, implica que tenemos un sesgo similar
# entonces lo que nos va a determinar acá todo es la varianza o que tan comprimido esta

# Por otro lado, el sesgo ( donde esta centrado ) despues lo puedo modificar para que se acerque a mi valor real de a1
# En cambio, la varianza no. Por lo que, me va a interar tener la menor varianza posible, ya que esto no lo puedo calibrar o mejorar de ninguna otra manera


# Índice correspondiente nuestro Omega_0
f_target = Omega_0
index_f_target = int(f_target / df)             # Con esto obtengo 250 hz

# Estimadores por ventana
estimadores = {}

for nombre, ventana in ventanas.items():
    
    # Normalización de la energía de la ventana
    win_rms = np.sqrt(np.mean(ventana**2))  
    Signal_windowed = Signal * ventana / win_rms
    
    # FFT normalizada
    X_f = fft(Signal_windowed, axis=0) / N  
    
    # Calculo estimador en nuestro Omega_0
    a1_estimado = np.abs(X_f[index_f_target, :])
    
    # Almaceno los estimadores
    estimadores[nombre] = a1_estimado

# Un solo gráfico con todos los histogramas
plt.figure(figsize=(12, 7))

for nombre, est in estimadores.items():
    plt.hist(est, bins=30, alpha=0.5, label=nombre, edgecolor='k', density=True)

# Líneas verticales: valor real y referencias
plt.axvline(A1, color='red', linestyle='--', label=r'Valor real $a_1 = \sqrt{2}$')

# También mostramos la media de cada estimador con una línea punteada verde
for nombre, est in estimadores.items():
    plt.axvline(np.mean(est), linestyle='--', label=f'Media {nombre}: {np.mean(est):.2f}')

plt.xlabel(r'Estimador $\hat{a}_1$')
plt.ylabel('Densidad')
plt.title('Distribución del estimador $\hat{a}_1$ para distintas ventanas')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
  
#%% Estimador de frecuencia: omega1 = argmax |X(omega)|

estimadores_omega1 = {}

for nombre, ventana in ventanas.items():
    
    # Normalizamos energía RMS
    win_rms = np.sqrt(np.mean(ventana**2))
    Signal_windowed = Signal * ventana / win_rms

    # FFT (normalizada)
    X_f = fft(Signal_windowed, axis=0)
    X_f = X_f[:N//2, :]  # Parte positiva

    # Índice de máxima magnitud para cada realizacion
    idx_max = np.argmax(np.abs(X_f), axis=0)    

    # Convertimos índices a frecuencia
    omega1_estimado = idx_max * df
    
    # Almacenamos estimadores
    estimadores_omega1[nombre] = omega1_estimado

#%% Histograma de los estimadores omega1

plt.figure(figsize=(12, 7))

for nombre, est in estimadores_omega1.items():
    plt.hist(est, bins=30, alpha=0.5, label=nombre, edgecolor='k', density=True)

# Línea del valor real de Omega_1 (el centro del intervalo)
plt.axvline(Omega_0, color='red', linestyle='--', label=r'Valor real $\Omega_0$')

# También mostramos la media de cada estimador con una línea punteada verde
for nombre, est in estimadores_omega1.items():
    plt.axvline(np.mean(est), linestyle='--', label=f'Media {nombre}: {np.mean(est):.2f} Hz')

plt.xlabel(r'Estimador $\hat{\omega}_1$ [Hz]')
plt.ylabel('Densidad')
plt.title('Distribución del estimador $\hat{\omega}_1$ para distintas ventanas')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#%%

# El estimador Omega 1 va a ser el sobre el modulo de la FFT de la señal Omega1 = argmax|X(omega)|
# Donde esa operacioj devuelve el valor de omega que maximiza el modulo de la TF

# Vemos que obtenemos que caso todo esta al rededor de 250 hz que es donde esta el maximo, pero al tener una res. espectral regular
# A lo sumo nos vamos un poco del 250

#%% Hacemos Cero Padding para aumentar la resolucion espectral y asi vemos mejor como son las variaciones de Omega1
```

```python
#%% Módulos a importar

import numpy as np
from scipy import signal
from scipy.signal.windows import hann, flattop, blackmanharris
from scipy.fft import fft
import matplotlib.pyplot as plt

#%% Función para correr simulación y devolver métricas

def run_simulation(SNR, fs=1000, N=1000, N_Test=200, Omega_0=None):
    ts = 1/fs
    df = fs/N
    if Omega_0 is None:
        Omega_0 = fs/4

    Sigma2 = 10**(-SNR/10)

    A1 = np.sqrt(2)

    fr = np.random.uniform(-1/2,1/2,N_Test).reshape(1,N_Test)
    Omega_1 = Omega_0 + fr*df

    tt = np.linspace(0, (N-1)*ts, N).reshape(N,1)
    tt = np.tile(tt, (1, N_Test))

    S = A1 * np.sin(2 * np.pi * Omega_1 * tt)

    SD_Sigma = np.sqrt(Sigma2)
    nn = np.random.normal(0, SD_Sigma, N).reshape(N,1)
    nn = np.tile(nn, (1,N_Test))

    Signal = S + nn

    ventanas = {
        'Rectangular': np.ones((N, 1)),
        'Hann': hann(N).reshape(N, 1),
        'FlatTop': flattop(N).reshape(N, 1),
        'Blackman-Harris': blackmanharris(N).reshape(N, 1)
    }

    df = fs/N
    f_target = Omega_0
    index_f_target = int(f_target / df)

    estimadores_a1 = {}
    estimadores_omega1 = {}

    for nombre, ventana in ventanas.items():
        win_rms = np.sqrt(np.mean(ventana**2))
        Signal_windowed = Signal * ventana / win_rms
        X_f = fft(Signal_windowed, axis=0) / N

        a1_estimado = np.abs(X_f[index_f_target, :])
        estimadores_a1[nombre] = a1_estimado

        X_f_pos = X_f[:N//2, :]
        idx_max = np.argmax(np.abs(X_f_pos), axis=0)
        omega1_estimado = idx_max * df
        estimadores_omega1[nombre] = omega1_estimado

    metricas_a1 = {}
    metricas_omega1 = {}

    for nombre, est in estimadores_a1.items():
        media = np.mean(est)
        varianza = np.var(est)
        sesgo = media - A1
        metricas_a1[nombre] = (media, sesgo, varianza)

    for nombre, est in estimadores_omega1.items():
        media = np.mean(est)
        varianza = np.var(est)
        sesgo = media - Omega_0
        metricas_omega1[nombre] = (media, sesgo, varianza)

    return metricas_a1, metricas_omega1

#%% Correr simulaciones para SNR = 10 y SNR = 3

a1_snr10, omega1_snr10 = run_simulation(SNR=10)
a1_snr3, omega1_snr3 = run_simulation(SNR=3)

#%% Mostrar comparativa de a1

print("\n" + "="*70)
print("Comparativa del estimador de amplitud a1")
print("="*70)
print(f"{'Ventana':<18}{'E[â1] (SNR=3)':>15}{'Sesgo':>12}{'Var':>12} | {'E[â1] (SNR=10)':>15}{'Sesgo':>12}{'Var':>12}")
print("-"*70)

for nombre in a1_snr3.keys():
    m3, s3, v3 = a1_snr3[nombre]
    m10, s10, v10 = a1_snr10[nombre]
    print(f"{nombre:<18}{m3:>15.4f}{s3:>12.4f}{v3:>12.4f} | {m10:>15.4f}{s10:>12.4f}{v10:>12.4f}")

#%% Mostrar comparativa de omega1

print("\n" + "="*70)
print("Comparativa del estimador de frecuencia omega1")
print("="*70)
print(f"{'Ventana':<18}{'E[ŵ1] (SNR=3)':>15}{'Sesgo':>12}{'Var':>12} | {'E[ŵ1] (SNR=10)':>15}{'Sesgo':>12}{'Var':>12}")
print("-"*70)

for nombre in omega1_snr3.keys():
    m3, s3, v3 = omega1_snr3[nombre]
    m10, s10, v10 = omega1_snr10[nombre]
    print(f"{nombre:<18}{m3:>15.4f}{s3:>12.4f}{v3:>12.4f} | {m10:>15.4f}{s10:>12.4f}{v10:>12.4f}")
```

### Analisis de lo observado en las pruebas

Empezando por los graficos empecemos por la representacion del espectro de la señal para cada ventana podemos observar que al estar tomando el promedio de los espectros de las 200 realizaciones para cada una de las ventanas en general el resultado obtenido es muy similar entre si, es decir, que no hay diferencias evidentes entre cada uno de las ventanas seleccionadas. Quiza podria remarcar que la ventana circuilar en turno al pico en $\Omega_0$ no es tan marcada en las inmediaciones del punto a comparacion de las demas ventanas, es decir, que el pico no se eleva tan rapido como en las demas ventanas. 

Adentremosnos en los estimadores. Comenzando por $a_1$, vemos que cada una de las ventanas tiene un sesgo diferente, es decir, que esta centrada en un punto diferente. Esto a priori no es un problema irreversible, ya que, llevando esto a un terreno mas real podriamos calibrar nuestro medidor para modificar en donde va a estar centrada la señal acercandola al valor requerido. Es decir, llevando el sesgo a 0 mediante una calibracion. Imaginemos que tenemos un tester y calibramos el mismo para asegurar que lo que estamos midiendo sea correcto. Ahora, en contraparte lo que mas nos interesa en este aspecto es que tanta varianza tenemos con cada ventana, debido a que las diferencias entre cada medicion no es algo que podamos corregir tal y como lo mencionamos con el sesgo, ya que depende del metodo que estamos utilizando. Con esto en cuenta, podriamos decir que la mejor ventana para el estimador a1 es la flattop que tiene la menor varianza. 

Luego si analizamos el estimador $\Omega_1$ podemos aplicar los mismos conceptos para elegir la ventana que tenga una menor varianza. Pero, algo que quiero remarcar que si bien en la gran mayoría los realizaciones tienen un $\Omega_1$ pegado a $\Omega_0$ pero esto se debe a la falta de resolucion que tenemos en este sentido, es decir, que van a caer en mas menos 1 bin al rededor de $\Omega_0$, y por eso vemos que todo cae en 249/250/251 hz. Ahora, si mejoramos la resolucion espectral de alguna manera vamos a ver que esto cambia. Lo trataremos mas adelante.

Ahora, algo que llama la atencion es que para un SNR = 3 tenemos mejores resoltados que para un SNR = 10. Como explicamos esto si en principio un SNR = 10 es mejor que uno igual a 3 ?

Si bien, un SNR = 10 implica que la potencia de nuestra señal es mucho mayor a la del ruido. Por lo que, la señal resultante es mas limpia. Al calcular los estimadores vemos que para $a_1$ se ve claramente que obtemos mejores resultados para un SNR = 10 que para uno igual a 3. Ahora, esto no pasa para $\Omega_1$ y como antes mencionamos, esto ocurre debido a la poca resolucion espectral que tenemos para la frecuencia en donde tenemos muchas frecuencias intermedias pero como solo hay 3 posibles bins esta variacion no es representable de manera fidedigna. ### 💎 Analice el efecto del zero-padding para el estimador Ω^1

Justamente, si hacemos Zero-padding nos ayuda a poder aumentar la resolucion espectral y asi tener mas valores intermedios para la frecuencia. Analicemos esto en un codigo:

```python
#%% módulos y funciones a importar

import numpy as np
from scipy import signal
from scipy.signal.windows import hamming, hann, blackman, kaiser, flattop, blackmanharris
from scipy.fft import fft, fftshift
import matplotlib as mpl
import matplotlib.pyplot as plt


#%% Genero mi Omega 1 que va a ser una frecuencia original mas una inceretidumbre en la frecuencia
# Esto genera que no sepa la frecuencia que estamos, sino que solo sepamos entre que valores va a estar

#%% Datos de la simulacion

fs = 1000           # frecuencia de muestreo (Hz)
N = 1000            # cantidad de muestras
ts = 1/fs           # tiempo de muestreo
df = fs/N           # resolución espectral

N_Test = 200        # Numero de pruebas

SNR = 10            # Signal to Noise Ratio 

Sigma2 = 10**-1     # Despejando el PN de la ecuacion de SNR llego a este valor

Omega_0 = fs/4      # Nos ponemos a mitad de banda digital

#%% Genero mi vector de 1000x200 de la senoidal

A1 = np.sqrt(2)                                             # Genero la amplitud de manera que el seno quede normalizado

fr = np.random.uniform(-1/2,1/2,N_Test).reshape(1,N_Test)   # Declaro mi vector 1x200 al pasar 200 valores a la uniforme     
                                                            # Fuerzo las dimensiones
Omega_1 = Omega_0 + fr*df                                   # Genero mi Omega 1

# Genero vector de tiempo para poder meterlo como mi vector de 1000 x 200 en el seno 
tt = np.linspace(0, (N-1)*ts, N).reshape(N,1)

tt = np.tile(tt, (1, N_Test))                               # Genero la matriz de 1000 x 200

# Al mutiplicar con * hacemos el producto matricial para que quede de 1000x200
S = A1 * np.sin(2 * np.pi * Omega_1 * tt)

varianzas = np.var(S, axis=0)                               # Varianza por cada una de las 200 señales
varianza_promedio = np.mean(varianzas)                      # Compruebo varianza 1 


#%% Genereo el ruido para la señal

# Para poder general la señal de ruido, tenemos que tener una distribucion normal con un N(o,sigma)

Media = 0                   # Media
SD_Sigma = np.sqrt(Sigma2)  # Desvio standar 

nn = np.random.normal(Media, SD_Sigma, N).reshape(N,1)              # Genero señal de ruido

nn = np.tile(nn, (1,N_Test))                                        # Ahora tengo que generaer mi matriz de ruido de 200x1000


#%% Ahora genero mi señal final sumando las matrices

# Esto seria mi x(k) = a1 * sen(W1 * k ) + n(k), siendo N(k) el ruido
Signal = S + nn

#%% Calculo las FFT de cada versión ( ventana, Hann, FlatTop, Blackman-Harris   )

#Declaro las ventanas que voy a usar y le doy N valores. Fuerzo dimensiones

ventanas = {
    'Rectangular': np.ones((N, 1)),
    'Hann': hann(N).reshape(N, 1),
    'FlatTop': flattop(N).reshape(N, 1),
    'Blackman-Harris': blackmanharris(N).reshape(N, 1)
}

num_senales = 200                                               # Cantidad de realizaciones
indices = np.linspace(0, N_Test - 1, num_senales, dtype=int)

# Eje de frecuencia solo parte positiva (0 a fs/2)
frec_pos = np.linspace(0, fs/2, N//2)

#%% Hacemos Cero Padding para aumentar la resolucion espectral y asi vemos mejor como son las variaciones de Omega1

estimadores_omega1_CeroPadding = {}

for nombre, ventana in ventanas.items():
    # Normalizamos energía 
    win_rms = np.sqrt(np.mean(ventana**2))
    Signal_windowed = Signal * ventana / win_rms

    # Agrego cantidad de ceros a agregar
    N_ZP = 5*N

    # FFT (normalizada)
    X_f = fft(Signal_windowed,N_ZP, axis=0) # FFT Con cero padding
    X_f = X_f[:N_ZP//2, :]                  # Parte positiva teniendo en cuenta el cambio de N

    # Índice de máxima magnitud para cada prueba
    idx_max = np.argmax(np.abs(X_f), axis=0)    
    
    # Recalculo la resolucion espectral
    df_ZP = fs / N_ZP

    # Convertimos índices a frecuencia
    omega1_estimado_CeroPadding = idx_max * df_ZP
    
    # Almaceno estimadores
    estimadores_omega1_CeroPadding [nombre] = omega1_estimado_CeroPadding 

#%% Repito cálculo para SNR = 3, recalculo todos los valores necesarios

SNR_3 = 3
Sigma2_3 = np.var(S) / (10**(SNR_3/10))
SD_Sigma_3 = np.sqrt(Sigma2_3)
nn_3 = np.random.normal(Media, SD_Sigma_3, (N, 1))
nn_3 = np.tile(nn_3, (1, N_Test))
Signal_3 = S + nn_3

estimadores_omega1_CeroPadding_SNR3 = {}

for nombre, ventana in ventanas.items():
    win_rms = np.sqrt(np.mean(ventana**2))
    Signal_windowed_3 = Signal_3 * ventana / win_rms

    N_ZP = 5 * N
    X_f_3 = fft(Signal_windowed_3, N_ZP, axis=0)
    X_f_3 = X_f_3[:N_ZP//2, :]

    idx_max_3 = np.argmax(np.abs(X_f_3), axis=0)
    df_ZP = fs / N_ZP
    omega1_estimado_CeroPadding_3 = idx_max_3 * df_ZP
    estimadores_omega1_CeroPadding_SNR3[nombre] = omega1_estimado_CeroPadding_3

#%% Subplot comparativo

fig, axs = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# SNR = 10
axs[0].set_title('SNR = 10 dB')
for nombre, est in estimadores_omega1_CeroPadding.items():
    axs[0].hist(est, bins=30, alpha=0.5, label=nombre, edgecolor='k', density=True)
    axs[0].axvline(np.mean(est), linestyle='--', label=f'Media {nombre}: {np.mean(est):.2f} Hz')
axs[0].axvline(Omega_0, color='red', linestyle='--', label=r'Valor real $\Omega_0$')
axs[0].set_ylabel('Densidad')
axs[0].legend()
axs[0].grid(True)

# SNR = 3
axs[1].set_title('SNR = 3 dB')
for nombre, est in estimadores_omega1_CeroPadding_SNR3.items():
    axs[1].hist(est, bins=30, alpha=0.5, label=nombre, edgecolor='k', density=True)
    axs[1].axvline(np.mean(est), linestyle='--', label=f'Media {nombre}: {np.mean(est):.2f} Hz')
axs[1].axvline(Omega_0, color='red', linestyle='--', label=r'Valor real $\Omega_0$')
axs[1].set_xlabel(r'Estimador $\hat{\omega}_1$ [Hz]')
axs[1].set_ylabel('Densidad')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()

#%% Tabla comparativa de sesgo y varianza

print("-" * 80)
print(f"{'Ventana':<20}{'E[ŵ1] SNR=10':>15}{'Sesgo':>10}{'Varianza':>12} | {'E[ŵ1] SNR=3':>15}{'Sesgo':>10}{'Varianza':>12}")
print("-" * 80)

for nombre in ventanas.keys():
    est_10 = estimadores_omega1_CeroPadding[nombre]
    est_3 = estimadores_omega1_CeroPadding_SNR3[nombre]

    media_10 = np.mean(est_10)
    sesgo_10 = media_10 - Omega_0
    varianza_10 = np.var(est_10)

    media_3 = np.mean(est_3)
    sesgo_3 = media_3 - Omega_0
    varianza_3 = np.var(est_3)

    print(f"{nombre:<20}{media_10:>15.2f}{sesgo_10:>10.2f}{varianza_10:>12.4f} | {media_3:>15.2f}{sesgo_3:>10.2f}{varianza_3:>12.4f}")
print("-" * 80)
```

### Que hacemos al hacer Zero-Padding ?

En pocas palabras estamos agregando N valores mas a la señal pero la particularidad es que todos estos estan en 0, por lo que no tenemos mas informacion en la señal sino que solamente mas puntos lo que nos ayuda a mejorar directamente la resolucion espectral. Esto lo podemos apreciar viendo que ahora en vez de solo tener los 3 Bins ( 249 / 250 / 251 hz ) tenemos muchos mas, por lo que, los valores que antes quedaban en un intermedio tienen mas posiciones en las cuales quedar con lo que tenemos un mejor analisis del estimador $\Omega_1$ y así tomar una mejor decision a la hora de elegir una ventana. 

Como podemos ver en este caso, no estamos agregando los ceros a la señal en este caso sino que directamente al hacer la FFT podemos indicar que haga el ZP para N valores que indicamos ( en este caso 5*N ), por lo que, no tenemos que hacer ninguna accion sobre la señal que medimos sino en este paso donde estamos representando a la señal agregamos directamente los ceros para mejorar la resolucion espectral antes mencionada. 

Ahora, como es el espectro de esta señal a la que le agregamos los ceros ?

```python
#%% módulos y funciones a importar

import numpy as np
from scipy import signal
from scipy.signal.windows import hamming, hann, blackman, kaiser, flattop, blackmanharris
from scipy.fft import fft, fftshift
import matplotlib as mpl
import matplotlib.pyplot as plt


#%% Genero mi Omega 1 que va a ser una frecuencia original mas una inceretidumbre en la frecuencia
# Esto genera que no sepa la frecuencia que estamos, sino que solo sepamos entre que valores va a estar

#%% Datos de la simulacion

fs = 1000           # frecuencia de muestreo (Hz)
N = 1000            # cantidad de muestras
ts = 1/fs           # tiempo de muestreo
df = fs/N           # resolución espectral

N_Test = 200        # Numero de pruebas

SNR = 10            # Signal to Noise Ratio 

Sigma2 = 10**-1     # Despejando el PN de la ecuacion de SNR llego a este valor

Omega_0 = fs/4      # Nos ponemos a mitad de banda digital

#%% Genero mi vector de 1000x200 de la senoidal

A1 = np.sqrt(2)                                             # Genero la amplitud de manera que el seno quede normalizado

fr = np.random.uniform(-1/2,1/2,N_Test).reshape(1,N_Test)   # Declaro mi vector 1x200 al pasar 200 valores a la uniforme     
                                                            # Fuerzo las dimensiones
Omega_1 = Omega_0 + fr*df                                   # Genero mi Omega 1

# Genero vector de tiempo para poder meterlo como mi vector de 1000 x 200 en el seno 
tt = np.linspace(0, (N-1)*ts, N).reshape(N,1)

tt = np.tile(tt, (1, N_Test))                               # Genero la matriz de 1000 x 200

# Al mutiplicar con * hacemos el producto matricial para que quede de 1000x200
S = A1 * np.sin(2 * np.pi * Omega_1 * tt)

varianzas = np.var(S, axis=0)                               # Varianza por cada una de las 200 señales
varianza_promedio = np.mean(varianzas)                      # Compruebo varianza 1 


#%% Genereo el ruido para la señal

# Para poder general la señal de ruido, tenemos que tener una distribucion normal con un N(o,sigma)

Media = 0                   # Media
SD_Sigma = np.sqrt(Sigma2)  # Desvio standar 

nn = np.random.normal(Media, SD_Sigma, N).reshape(N,1)              # Genero señal de ruido

nn = np.tile(nn, (1,N_Test))                                        # Ahora tengo que generaer mi matriz de ruido de 200x1000


#%% Ahora genero mi señal final sumando las matrices

# Esto seria mi x(k) = a1 * sen(W1 * k ) + n(k), siendo N(k) el ruido
Signal = S + nn

#%% Calculo las FFT de cada versión ( ventana, Hann, FlatTop, Blackman-Harris   )

#Declaro las ventanas que voy a usar y le doy N valores. Fuerzo dimensiones

ventanas = {
    'Rectangular': np.ones((N, 1)),
    'Hann': hann(N).reshape(N, 1),
    'FlatTop': flattop(N).reshape(N, 1),
    'Blackman-Harris': blackmanharris(N).reshape(N, 1)
}

num_senales = 200                                               # Cantidad de realizaciones
indices = np.linspace(0, N_Test - 1, num_senales, dtype=int)

# Eje de frecuencia solo parte positiva (0 a fs/2)
frec_pos = np.linspace(0, fs/2, N//2)

#%% Hacemos Cero Padding para aumentar la resolucion espectral y asi vemos mejor como son las variaciones de Omega1

estimadores_omega1_CeroPadding = {}

for nombre, ventana in ventanas.items():
    # Normalizamos energía 
    win_rms = np.sqrt(np.mean(ventana**2))
    Signal_windowed = Signal * ventana / win_rms

    # Agrego cantidad de ceros a agregar
    N_ZP = 5*N

    # FFT (normalizada)
    X_f = fft(Signal_windowed,N_ZP, axis=0) # FFT Con cero padding
    X_f = X_f[:N_ZP//2, :]                  # Parte positiva teniendo en cuenta el cambio de N

    # Índice de máxima magnitud para cada prueba
    idx_max = np.argmax(np.abs(X_f), axis=0)    
    
    # Recalculo la resolucion espectral
    df_ZP = fs / N_ZP

    # Convertimos índices a frecuencia
    omega1_estimado_CeroPadding = idx_max * df_ZP
    
    # Almaceno estimadores
    estimadores_omega1_CeroPadding [nombre] = omega1_estimado_CeroPadding 

#%% Señal con ventana Hann (ya normalizada en energía RMS)
ventana = ventanas['Hann']
win_rms = np.sqrt(np.mean(ventana**2))
Signal_windowed = Signal * ventana / win_rms

# FFT sin zero padding
X_f_noZP = fft(Signal_windowed, axis=0)
X_f_noZP = X_f_noZP[:N//2, :]
f_noZP = np.linspace(0, fs/2, N//2)
mag_noZP_dB = 20 * np.log10(np.mean(np.abs(X_f_noZP), axis=1) + 1e-12)  # Evitar log(0)

# FFT con zero padding
N_ZP = 5 * N
X_f_ZP = fft(Signal_windowed, N_ZP, axis=0)
X_f_ZP = X_f_ZP[:N_ZP//2, :]
f_ZP = np.linspace(0, fs/2, N_ZP//2)
mag_ZP_dB = 20 * np.log10(np.mean(np.abs(X_f_ZP), axis=1) + 1e-12)

# Gráfico en dB
plt.figure(figsize=(12, 6))
plt.plot(f_noZP, mag_noZP_dB, label='Sin Zero Padding', lw=2)
plt.plot(f_ZP, mag_ZP_dB, label='Con Zero Padding (5N)', lw=2, linestyle='--')
plt.axvline(Omega_0, color='red', linestyle=':', label=r'$\Omega_0$ Real')

plt.title('Espectro promedio (ventana Hann) con y sin Zero Padding [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud [dB]')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Gráfico comparativo mejorado
plt.figure(figsize=(12, 6))

# Espectros promedio
plt.plot(f_noZP, mag_noZP_dB, label='Sin Zero Padding', color='C0', lw=2)
plt.plot(f_ZP, mag_ZP_dB, label='Con Zero Padding (5N)', color='C1', lw=2, linestyle='--')

# Línea de la frecuencia real
plt.axvline(Omega_0, color='red', linestyle=':', lw=2, label=r'$\Omega_0$ Real')

# Zoom en zona de interés (±30 Hz alrededor de Omega_0)
plt.xlim(Omega_0 - 30, Omega_0 + 30)

# Ajustes estéticos
plt.title('Comparación del espectro promedio con y sin Zero Padding (ventana Hann)', fontsize=14)
plt.xlabel('Frecuencia [Hz]', fontsize=12)
plt.ylabel('Magnitud [dB]', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.tight_layout()

# Anotación visual
plt.annotate('Mayor resolución\nfrecuencial',
             xy=(Omega_0 + 10, mag_ZP_dB[np.argmin(np.abs(f_ZP - (Omega_0 + 10)))]),
             xytext=(Omega_0 + 15, -20),
             arrowprops=dict(arrowstyle="->", lw=1.5),
             fontsize=11)

plt.show()
```

### 🤯 Proponga estimadores alternativos para frecuencia y amplitud de la senoidal y repita el experimento.

Vamos a proponer un estimador usando un ajuste de curvas con Curve Fit

Para eso vamos a plantear un modelo en el cual es a lo que queremos llegar, que en este caso es una señal sinosuidal a la cual le vamos a dar algunos parametros de entrada. Luego, con esto ya seteado usamos curve fit donde se trata de encontrar los valores de a y f que minimizan el error cuadrático entre el modelo y los datos. Internamente usa Levenberg-Marquardt, un método iterativo que combina descenso del gradiente y Newton-Raphson para ajustar funciones no lineales.

Luego lo vamos a repetir 200 veces para cada una de las ventanas y de esa manera confeccionamos todo lo necesario. 

Lo remarcable de este metodo es que estamos calculando ambos estimadores al mismo tiempo de manera que estamos minimizando los errores que podamos cometer en el calculo y a su vez el calculo de ambos esta relacionado. Algo que no me cierra tanto es que necesitamos saber como es la señal que tenemos, es decir, que de antemano tengo que saber el tipo de señal que tengo para poder usar este metodo y en los estimadores anteriores que teniamos dependia un poco mas de los parametros de la simulacion los cuales creo que es mas facil de saber a la forma de la señal que estamos midiendo 

```python
import numpy as np
from scipy import signal
from scipy.signal.windows import hann, flattop, blackmanharris
from scipy.fft import fft
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Parámetros de simulación
fs = 1000
N = 1000
ts = 1/fs
df = fs/N
N_Test = 200
SNR = 10
Sigma2 = 10**-1
Omega_0 = fs/4
A1 = np.sqrt(2)

# Generacion de señal simplificado
fr = np.random.uniform(-1/2, 1/2, N_Test).reshape(1, N_Test)
Omega_1 = Omega_0 + fr * df
tt = np.linspace(0, (N-1)*ts, N).reshape(N, 1)
tt = np.tile(tt, (1, N_Test))
S = A1 * np.sin(2 * np.pi * Omega_1 * tt)
nn = np.random.normal(0, np.sqrt(Sigma2), (N, 1))
nn = np.tile(nn, (1, N_Test))
Signal = S + nn

#Declaro las ventanas que voy a usar
ventanas = {
    'Rectangular': np.ones((N, 1)),
    'Hann': hann(N).reshape(N, 1),
    'FlatTop': flattop(N).reshape(N, 1),
    'Blackman-Harris': blackmanharris(N).reshape(N, 1)
}

# Modelo senoidal para ajuste, esto va a ser a lo que voy a querer que se parezca mi salida por lo que es con lo que voy a comparar. 
def modelo_senoidal(t, a, f, phi):
    return a * np.sin(2 * np.pi * f * t + phi)

# Estimadores usando curve_fit
estimadores_curvefit = {}

for nombre, ventana in ventanas.items():
    print(f"Estimando con ventana: {nombre}")

    # Ventana con energia normalizada
    win_rms = np.sqrt(np.mean(ventana**2))
    Signal_windowed = Signal * ventana / win_rms

    #Acomulamos los estimadores que vamos calculando en arrays
    a1_estimados = []
    w1_estimados = []

    # Esto ya es parte del codigo que encontre de como implementar esto. En pocas palabras estamos tratando de ajustar la señal ventaneada que tenemos
    # al comparar los parametros del modelo de señal que hicimos con una de las 200 señales que tenemos para cada ventana
    
    for i in range(N_Test):
        xk = Signal_windowed[:, i]
        try:
            popt, _ = curve_fit(modelo_senoidal, tt[:, i], xk, p0=[1.0, Omega_0, 0], maxfev=2000)
            a1_estimados.append(popt[0])
            w1_estimados.append(popt[1])
        except:
            a1_estimados.append(np.nan)
            w1_estimados.append(np.nan)
    
    estimadores_curvefit[nombre] = {
        'a1': np.array(a1_estimados),
        'w1': np.array(w1_estimados)
    }

# Gráficos: Subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, (nombre, datos) in enumerate(estimadores_curvefit.items()):
    axs[i].hist(datos['a1'], bins=30, alpha=0.6, label=r'$\hat{a}_1$', color='skyblue', edgecolor='k')
    axs[i].axvline(A1, color='r', linestyle='--', label='Valor real $a_1$')
    axs[i].set_title(f'Estimador $a_1$ - {nombre}')
    axs[i].set_xlabel('$\hat{a}_1$')
    axs[i].set_ylabel('Frecuencia')
    axs[i].grid(True)
    axs[i].legend()

plt.tight_layout()
plt.show()

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, (nombre, datos) in enumerate(estimadores_curvefit.items()):
    axs[i].hist(datos['w1'], bins=30, alpha=0.6, label=r'$\hat{\omega}_1$', color='lightgreen', edgecolor='k')
    axs[i].axvline(Omega_0, color='r', linestyle='--', label='Valor real $\omega_0$')
    axs[i].set_title(f'Estimador $\omega_1$ - {nombre}')
    axs[i].set_xlabel('$\hat{\omega}_1$ [Hz]')
    axs[i].set_ylabel('Frecuencia')
    axs[i].grid(True)
    axs[i].legend()

plt.tight_layout()
plt.show()

# Tabla comparativa de sesgo y varianza
print("-" * 70)
print(f"{'Ventana':<20}{'E[a1]':>10}{'Sesgo a1':>12}{'Var[a1]':>12}{'Sesgo w1':>12}{'Var[w1]':>12}")
print("-" * 70)

for nombre, datos in estimadores_curvefit.items():
    a1 = datos['a1']
    w1 = datos['w1']
    a1 = a1[~np.isnan(a1)]
    w1 = w1[~np.isnan(w1)]
    
    media_a1 = np.mean(a1)
    var_a1 = np.var(a1)
    sesgo_a1 = media_a1 - A1
    
    media_w1 = np.mean(w1)
    var_w1 = np.var(w1)
    sesgo_w1 = media_w1 - Omega_0
    
    print(f"{nombre:<20}{media_a1:>10.4f}{sesgo_a1:>12.4f}{var_a1:>12.4f}{sesgo_w1:>12.4f}{var_w1:>12.4f}")
```

Con este metodo de estimación, podriamos observar que en este caso la Flattop tiene una mejor varianza tanto en A1 como en W1, de manera que podria ser nuestra elección de ventana para realizar este experimento

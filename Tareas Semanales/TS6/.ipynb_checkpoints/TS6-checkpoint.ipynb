Por otro lado, tenemos que tener en cuenta que al operar de dicha manera vamos a obtener Hr(w), lo cual no es la T(z) pero que si el valor en modulo de ambos coincide. De esta manera, podemos obtener la respuesta en modulo de T(z).

## Hallar T(z)=Y(z)X(z) / Calcular su respuesta en frecuencia de módulo y fase.

Observacion: Vamos a hacer en detalle el punto A, y luego como la operatoria de los demas es muy similar vamos a exponer los resultados encontrados. 

Para comenzar vamos a utilizar la transformada bilateral para pasar al dominio de la transformada zeta y mas particularmente vamos a utilizar la propiedad de la demora que estipulaba que:

$$
\mathcal{Z}\{x[n - n_0]\} = z^{-n_0} X(z)
$$

Con lo que, utilizando esta propiedad en la expresion del punto a, obtenemos:

$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n) \xrightarrow{\mathcal{Z}} Y(z) = z^{-3}X(z) + z^{-2}X(z) + z^{-1}X(z) + X(z)
$$

Ahora, sacando factor común X(z):

$$
Y(z) = X(z) (z^{-3} + z^{-2} + z^{-1} + 1 ) \rightarrow  T(z) = z^{-3} + z^{-2} + z^{-1} + 1 
$$

Ya con la T(z) calculada queremos trabajar dicha espresion para tenerla de manera mas comoda eliminando los exponentes negativos:

$$
T(z) = \frac{z^3 + z^2 + z + 1}{z^3}
$$

Ya con esta expresion podemos usar lo antes mencionado, al reemplazar $ Z = e^{jw} $:

$$
T(e^{j\omega}) = \frac{e^{j3\omega} + e^{j2\omega} + e^{j\omega} + 1}{e^{j3\omega}}
$$

Algo fundamental es pode llevar esta expresion a algo que quede en funcion de senos y/o cosenos para así obtener el Hr(w). Con este fin, vamos a sacar el factor común de los exponentes del numerador y del denominador. De esta manera, vamos a llevar ambas expresiones a terminos que podemos utilizar las identidades trigonometricas de los numeros complejos para reemplazar por senos/cosenos:

$$
T(e^{j\omega}) = \left( \frac{e^{j\frac{3\omega}{2}}}{e^{j3\omega}} \right) \cdot \left( e^{j\frac{3\omega}{2}} + e^{j\frac{\omega}{2}} + e^{-j\frac{\omega}{2}} + e^{-j\frac{3\omega}{2}} \right) \rightarrow T(e^{j\omega}) = e^{-j\frac{3\omega}{2}} \left( e^{j\frac{3\omega}{2}} + e^{j\frac{\omega}{2}} + e^{-j\frac{\omega}{2}} + e^{-j\frac{3\omega}{2}} \right) \rightarrow T(e^{j\omega}) = e^{-j\frac{3\omega}{2}} \left[ 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right]
$$

De esta manera tenemos un numero complejo expresado de forma exponencial, del cual el modulo va a ser Hr(w), con el cual podemos obtener:

$$
\left| H_R(\omega) \right| = \left| 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right| \quad \xrightarrow{} \quad \left| H_R(\omega) \right| = |T(j\omega)|
$$

$$
\quad \angle T(j\omega) = \angle e^{-j\frac{3\omega}{2}} = -\frac{3\omega}{2}
$$

Resumiendo, podemos decir que para la expresion:

$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n)
$$

$$
\left\{
\begin{array}{l}
\displaystyle
T(z) = \frac{z^3 + z^2 + z + 1}{z^3} \\[10pt]
\displaystyle
\left| T(j\omega) \right| = \left| 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right| \\[10pt]
\displaystyle
\angle T(j\omega) = \angle e^{-j\frac{3\omega}{2}} = -\frac{3\omega}{2}
\end{array}
\right.
$$




Ahora, con todo esto hecho paso a paso vamos a hacer lo mismo para cada uno de los incisos restantes:

$$
\text{b) } y(n) = x(n-4) + x(n-3) + x(n-2) + x(n-1) + x(n)
$$

$$
\left\{
\begin{array}{l@{\quad}} 
\displaystyle
T(z) = \frac{1 + z + z^2 + z^3 + z^4}{z^4} \\[12pt]
\displaystyle
T(j\omega) = e^{-j2\omega} \left(1 + e^{j\omega} + e^{j2\omega} + e^{j3\omega} + e^{j4\omega}\right) = e^{-j2\omega} \left[2\cos(2\omega) + 2\cos(\omega) + 1\right] \\[12pt]
\displaystyle
\left| T(j\omega) \right| = \left| 2\cos(2\omega) + 2\cos(\omega) + 1 \right| \\[12pt]
\displaystyle
\angle T(j\omega) = \angle e^{-j2\omega} = -2\omega
\end{array}
\right.
$$

$$ \\ $$

$$
\text{c) } y(n) = x(n) - x(n-1)
$$

$$
\left\{
\begin{array}{l@{\quad}}
\displaystyle
T(z) = \frac{z - 1}{z} \\[12pt]
\displaystyle
T(j\omega) = e^{-j\frac{\omega}{2}} \left(e^{j\frac{\omega}{2}} - e^{-j\frac{\omega}{2}}\right) = e^{-j\frac{\omega}{2}} \cdot 2j \sin\left(\frac{\omega}{2}\right) \\[12pt]
\displaystyle
\left| T(j\omega) \right| = \left| 2 \sin\left(\frac{\omega}{2}\right) \right| \\[12pt]
\displaystyle
\angle T(j\omega) = \angle e^{-j\frac{\omega}{2}} + \angle \left(2j \sin\left(\frac{\omega}{2}\right)\right) = -\frac{\omega}{2} + \frac{\pi}{2}
\end{array}
\right.
$$

$$ \\ $$

$$
\text{d) } y(n) = x(n) - x(n-2)
$$

$$
\left\{
\begin{array}{l@{\quad}}
\displaystyle
T(z) = \frac{z^2 - 1}{z^2} \\[12pt]
\displaystyle
T(j\omega) = e^{-j\omega} \left(e^{j\omega} - e^{-j\omega}\right) = e^{-j\omega} \cdot 2j \sin(\omega) \\[12pt]
\displaystyle
\left| T(j\omega) \right| = \left| 2 \sin(\omega) \right| \\[12pt]
\displaystyle
\angle T(j\omega) = \angle e^{-j\omega} + \angle \left(2j \sin(\omega)\right) = -\omega + \frac{\pi}{2}
\end{array}
\right.
$$


Ahora, con todo esto vamos a generar un script de python para corroborar lo obtenido y a su vez poder graficarlo.

Disclaimer, 

Esta parte del codigo la hice con ayuda de la IA. El funcionamiento del codigo es el siguiente; Generamos una funcion para cada uno de los casos, esta funcion opera de la siguiente manera:

Defino lo obtenido con el analisis de los filtros de media movil, esos son mis valores teoricos
Calculo con :  
* w_vals, h = signal.freqz(b_coeffs, a, worN=1024)
* h_mag = np.abs(h)
* h_phase = np.angle(h)
El valor de T(z), su modulo y fase. Pero hacemos una salvedad para que si el analisis incluye senos o cosenos hagamos el las cuentas de una manera que lo aproximemos correctamente, por ese motivo tenemos el if que diferencia entre senos / Cosenos
Grafico lo obtenido en un subplot contrastando lo teorico con lo calculado 

Con todo esto en mente, tenemos un codigo que compara lo teorico con lo obtenido numericamente con numpy.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import sympy as sp
from collections import Counter
from IPython.display import display, Math

def analizar_sistema(b_coeffs, titulo, use_cos=True):
    # =============================
    # 1. Definición simbólica
    # =============================

    z, w = sp.symbols('z w', complex=True)

    N = len(b_coeffs)
    numerador = sum([b_coeffs[k] * z**(N-1 - k) for k in range(N)])
    denominador = z**(N-1)
    Tz = numerador / denominador
    display(Math(r"\text{" + titulo + r"}"))
    display(Math(r"T(z) = " + sp.latex(Tz)))

    ejw = sp.exp(sp.I * w)
    Tejw = Tz.subs(z, ejw)

    ejw_factor = sp.exp(sp.I * (N-1) * w / 2)
    A_w = Tejw * ejw_factor
    A_w_simplified = sp.simplify(sp.expand_trig(A_w))

    display(Math(r"T(e^{j\omega}) = e^{-j\frac{" + str(N-1) + r"\omega}{2}} \cdot A(\omega)"))
    display(Math(r"A(\omega) = " + sp.latex(A_w_simplified)))

    display(Math(r"|T(e^{j\omega})| = \left| A(\omega) \right|"))
    display(Math(r"\angle T(e^{j\omega}) = -\frac{" + str(N-1) + r"\omega}{2}"))

    # =============================
    # 2. Cálculo numérico con SciPy / NumPy
    # =============================

    a = [1]

    w_vals, h = signal.freqz(b_coeffs, a, worN=1024)
    h_mag = np.abs(h)
    h_phase = np.angle(h)

    if use_cos:
        def A_theory(w):
            val = 0
            for k in range(N):
                val += b_coeffs[k] * np.cos(w * ((N-1)/2 - k))
            return val
        A_vals = A_theory(w_vals)
    else:
        def A_exact(w):
            ejw = np.exp(1j * w)
            numerator = sum([b_coeffs[k] * ejw ** (N - 1 - k) for k in range(N)])
            denominator = ejw ** (N - 1)
            T_ejw = numerator / denominator
            A_w = T_ejw * np.exp(1j * (N - 1) / 2 * w)
            return np.abs(A_w)
        A_vals = A_exact(w_vals)

    phase_theory = - (N-1) / 2 * w_vals

    # =============================
    # 3. Gráficos
    # =============================

    fig, axs = plt.subplots(1, 3, figsize=(18, 4))

    # Ceros
    zeros = np.roots(b_coeffs)

    # Polos: como denominador es z^(N-1), hay N-1 polos en 0
    poles = np.array([0] * (N-1)) if N > 1 else np.array([])

    # Función para contar multiplicidades
    def contar_mult(puntos, tol=1e-5):
        if len(puntos) == 0:
            return {}
        puntos_redondeados = []
        for p in puntos:
            r = round(p.real / tol) * tol
            i = round(p.imag / tol) * tol
            puntos_redondeados.append(r + 1j * i)
        cuenta = Counter(puntos_redondeados)
        return dict(cuenta)

    mult_zeros = contar_mult(zeros)
    mult_poles = contar_mult(poles)

    theta = np.linspace(0, 2*np.pi, 500)
    axs[0].plot(np.cos(theta), np.sin(theta), 'k--', alpha=0.5)  # Circunferencia unitaria

    # Graficar ceros con multiplicidad
    for punto, mult in mult_zeros.items():
        axs[0].scatter(punto.real, punto.imag, s=80, marker='o', facecolors='none', edgecolors='b')
        label = "Cero"
        if mult > 1:
            label += f" mult. {mult}"
        axs[0].text(punto.real + 0.05, punto.imag + 0.05, label, color='b')

    # Graficar polos con multiplicidad
    for punto, mult in mult_poles.items():
        axs[0].scatter(punto.real, punto.imag, s=80, marker='x', color='r')
        label = "Polo"
        if mult > 1:
            label += f" mult. {mult}"
        axs[0].text(punto.real + 0.05, punto.imag + 0.05, label, color='r')

    axs[0].set_title('Diagrama de Polos y Ceros')
    axs[0].set_xlabel('Re')
    axs[0].set_ylabel('Im')
    axs[0].grid()
    axs[0].set_aspect('equal', adjustable='datalim')
    axs[0].set_xlim([-1.5, 1.5])
    axs[0].set_ylim([-1.5, 1.5])

    # Módulo
    axs[1].plot(w_vals, h_mag, label='|T(e^{jω})| Numérico', linewidth=2)
    axs[1].plot(w_vals, np.abs(A_vals), '--', label='|A(ω)| Teórico', linewidth=2)
    axs[1].set_title('Comparación del Módulo')
    axs[1].set_xlabel('ω [rad/muestra]')
    axs[1].set_ylabel('Módulo')
    axs[1].legend()
    axs[1].grid()

    # Fase
    axs[2].plot(w_vals, h_phase, label='Fase Numérica', linewidth=2)
    axs[2].plot(w_vals, phase_theory, '--', label=f'Fase Teórica: -{(N-1)/2}ω', linewidth=2)
    axs[2].set_title('Comparación de la Fase')
    axs[2].set_xlabel('ω [rad/muestra]')
    axs[2].set_ylabel('Fase [rad]')
    axs[2].legend()
    axs[2].grid()

    plt.tight_layout()
    plt.show()

# ======= Casos =======
b_a = [1, 1, 1, 1]
analizar_sistema(b_a, "Caso a) y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)")

b_b = [1, 1, 1, 1, 1]
analizar_sistema(b_b, "Caso b) y(n)=x(n−4)+x(n−3)+x(n−2)+x(n−1)+x(n)")

b_c = [1, -1]
analizar_sistema(b_c, "Caso c) y(n)=x(n)−x(n−1)", use_cos=False)

b_d = [1, 0, -1]
analizar_sistema(b_d, "Caso d) y(n)=x(n)−x(n−2)", use_cos=False)
```

### Conclusiones

Luego de ver el codigo pudimos corroborar que lo obtenido teoricamente coincide con lo que se obtiene del analisis con metodos numericos. 

Como primera impresion podemos ver que en todos los casos tenemos saltos en la fase, esto hace referencia a que cuando recorremos la circunferencia de 0 a Pi y pasamos por un Cero es cuando ocurre dicho salto en la fase. Por otro lado, por mas que tenemos un salto la pendiente que tiene la fase sigue siendo la misma que la que calculamos, por lo que, a pesar de tener este salto sigue la misma pendiente que lo calculado teoricamente.

Me resulta interesante que en todos los casos el modulo de la transferencia que calculamos teoricamente es igual a lo obtenido numericamente. Para mi, esto es algo que indica la robuestes del metodo por el cual haciendo calculos simples podemos llegar a los mismos resultados que usando un metodo numerico que requiere de poder computacional. Es decir, que calculando manualmente llegamos a lo mismo que con los metodos de manera que podemos comprobar rapidamente lo obtenido. Eso lo veo beneficioso a la hora de diseñar este tipo de filtros. 

Aunque la función de transferencia de estos filtros incluye un denominador con potencias de \( z \), por ejemplo:

$$
T(z) = \frac{z^{N-1} + \dots + 1}{z^{N-1}},
$$

este denominador no representa retroalimentación, sino simplemente un retraso en la señal. Por lo tanto:

- La salida depende únicamente de una combinación lineal finita de entradas actuales y pasadas.
- No hay dependencia de salidas anteriores (no hay realimentación).
- La respuesta al impulso es finita debido a esto anterior.

Esto implica que estos filtros son **Filtros FIR (Finite Impulse Response)**, a pesar de tener un término en el denominador que solo representa un retardo.

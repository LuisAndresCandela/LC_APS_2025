## Hallar T(z)=Y(z)X(z) / Calcular su respuesta en frecuencia de módulo y fase.

Observacion: Vamos a hacer en detalle el punto A, y luego como la operatoria de los demas es muy similar vamos a exponer los resultados encontrados. 

Para comenzar vamos a utilizar la transformada bilateral para pasar al dominio de la transformada zeta y mas particularmente vamos a utilizar la propiedad de la demora que estipulaba que:

$$
\mathcal{Z}\{x[n - n_0]\} = z^{-n_0} X(z)
$$

Con lo que, utilizando esta propiedad en la expresion del punto a, obtenemos:

$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n) \xrightarrow{\mathcal{Z}} Y(z) = z^{-3}X(z) + z^{-2}X(z) + z^{-1}X(z) + X(z)
$$

Ahora, sacando factor común X(z):

$$
Y(z) = X(z) (z^{-3} + z^{-2} + z^{-1} + 1 ) \rightarrow  T(z) = z^{-3} + z^{-2} + z^{-1} + 1 
$$

Ya con la T(z) calculada queremos trabajar dicha espresion para tenerla de manera mas comoda eliminando los exponentes negativos:

$$
T(z) = \frac{z^3 + z^2 + z + 1}{z^3}
$$

Ya con esta expresion podemos usar lo antes mencionado, al reemplazar $ Z = e^{jw} $:

$$
T(e^{j\omega}) = \frac{e^{j3\omega} + e^{j2\omega} + e^{j\omega} + 1}{e^{j3\omega}}
$$

Algo fundamental es pode llevar esta expresion a algo que quede en funcion de senos y/o cosenos para así obtener el Hr(w). Con este fin, vamos a sacar el factor común de los exponentes del numerador y del denominador. De esta manera, vamos a llevar ambas expresiones a terminos que podemos utilizar las identidades trigonometricas de los numeros complejos para reemplazar por senos/cosenos:

$$
T(e^{j\omega}) = \left( \frac{e^{j\frac{3\omega}{2}}}{e^{j3\omega}} \right) \cdot \left( e^{j\frac{3\omega}{2}} + e^{j\frac{\omega}{2}} + e^{-j\frac{\omega}{2}} + e^{-j\frac{3\omega}{2}} \right) \rightarrow T(e^{j\omega}) = e^{-j\frac{3\omega}{2}} \left( e^{j\frac{3\omega}{2}} + e^{j\frac{\omega}{2}} + e^{-j\frac{\omega}{2}} + e^{-j\frac{3\omega}{2}} \right) \rightarrow T(e^{j\omega}) = e^{-j\frac{3\omega}{2}} \left[ 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right]
$$

De esta manera tenemos un numero complejo expresado de forma exponencial, del cual el modulo va a ser Hr(w), con el cual podemos obtener:

$$
\left| H_R(\omega) \right| = \left| 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right| \quad \xrightarrow{} \quad \left| H_R(\omega) \right| = |T(j\omega)|
$$

$$
\quad \angle T(j\omega) = \angle e^{-j\frac{3\omega}{2}} = -\frac{3\omega}{2}
$$

Resumiendo, podemos decir que para la expresion:

$$
y(n) = x(n-3) + x(n-2) + x(n-1) + x(n)
$$

$$
\left\{
\begin{array}{l}
\displaystyle
T(z) = \frac{z^3 + z^2 + z + 1}{z^3} \\[10pt]
\displaystyle
\left| T(j\omega) \right| = \left| 2\cos\left(\frac{3\omega}{2}\right) + 2\cos\left(\frac{\omega}{2}\right) \right| \\[10pt]
\displaystyle
\angle T(j\omega) = \angle e^{-j\frac{3\omega}{2}} = -\frac{3\omega}{2}
\end{array}
\right.
$$




Ahora, con todo esto hecho paso a paso vamos a hacer lo mismo para cada uno de los incisos restantes:

\text{b) } y(n) = x(n-4) + x(n-3) + x(n-2) + x(n-1) + x(n)
